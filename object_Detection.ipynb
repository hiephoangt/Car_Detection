```python
import os
import cv2
import matplotlib.pyplot as plt

from ultralytics import YOLO
```

## Training

```python
# Load a model
yolo_yaml_path = 'yolo_data/data.yml'
model = YOLO('yolov8s.yaml').load('yolov8s.pt')
```

```python
epochs = 15
imgsz = 640
batch_size = 8
lr = 0.0005
patience = 5

results = model.train(
    data = yolo_yaml_path,
    epochs = epochs,
    imgsz = imgsz,
    batch = batch_size,
    lr0 = lr,
    patience =patience,
    project = 'models',
    name = 'yolov8/detect/train'
)
```

## Predict

```python
test_img_path ='/content/yolo_data/val/images/vid_4_12140.jpg'
model_path ='/content/models/yolov8/detect/train/weights/best.pt'

model = YOLO(model_path)
model.predict(test_img_path, save = True)
```

## Predict for many sample

```python
test_image_dir='/content/yolo_data/val/images'
for image_name in os.listdir('/content/yolo_data/val/images'):
  image_path = os.path.join(test_image_dir, image_name)
  model.predict(image_path,save = True)
```
